```python
paths = get_folders('data')
paths
```

```python
path = paths[0]
```

```python
datas = pd.read_csv(path+'exif_row.csv')
datas
```

```python
# Getting the mean values of the GPS to locate the center of the map to be plot

meanLat = datas.Latitude.mean()
meanLon = datas.Longitude.mean()

# Start a map

m = folium.Map([meanLat,meanLon],zoom_start=25)

# Plot markers showing the first and last photos

folium.Marker(location=[datas.iloc[[0]].Latitude,datas.iloc[[0]].Longitude],popup='Start',
              icon=folium.Icon(color='green')).add_to(m)
folium.Marker(location=[datas.iloc[[-1]].Latitude,datas.iloc[[-1]].Longitude],popup='Finish',
              icon=folium.Icon(color='red')).add_to(m)

# Plot circles for each of the GPS coordinates (each color is one different row)

for d in datas.iterrows():
    c = colors[d[1].Row-1]
    text = d[1].File+'\n'+d[1].TimeStamp+'\n'+str(d[1].Height)+'\n'+d[1].Dir
    folium.Circle(radius=1,location=[d[1].Latitude,d[1].Longitude],popup=text,color=c).add_to(m)
```
